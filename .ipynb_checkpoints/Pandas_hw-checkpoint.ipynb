```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(10)
#school_data_complete.groupby(0).count()
```

```python
Total_Schools=school_data_complete["school_name"].unique().size
Total_Students=school_data_complete["student_name"].size
Total_Budget="${:,.2f}".format(school_data_complete["budget"].unique().sum())
Average_Math_Score=school_data_complete["math_score"].mean()
Average_Reading_Score=school_data_complete["reading_score"].mean()
Overall_Passing=(Average_Math_Score+Average_Reading_Score)/2
Passing_Math1=school_data_complete.loc[school_data_complete["math_score"]>=70,"math_score"]
Passing_Math=Passing_Math1.size*100/Total_Students
Passing_Reading1=school_data_complete.loc[school_data_complete["reading_score"]>=70,"reading_score"]
Passing_Reading=Passing_Reading1.size*100/Total_Students

summary=pd.DataFrame({"Total Schools":[Total_Schools], 
                     "Total Students":['{:,.0f}'.format(Total_Students)],
                     "Total Budget":[Total_Budget],
                     "Average Math Score":[Average_Math_Score],
                     "Average Reading Score":[Average_Reading_Score],
                     "% Passing Math":[Passing_Math],
                     "% Passing Reading":[Passing_Reading],
                     "% Overall Passing Rate":[Overall_Passing]})
summary
```

```python
dataadd=school_data_complete
dataadd["Pass Math"]=Passing_Math1
dataadd["Pass Reading"]=Passing_Reading1
Schools=dataadd.groupby(["school_name"])

School_type=Schools["type"].unique()
Total_Students_g=Schools["student_name"].count()
Total_Budget_g=Schools["budget"].max()
Student_Budget_g=Total_Budget_g/Total_Students_g
Average_Math_Score_g=Schools["math_score"].mean()
Average_Reading_Score_g=Schools["reading_score"].mean()
Passing_Math_g=Schools["Pass Math"].count()*100/Total_Students_g
Passing_Reading_g=Schools["Pass Reading"].count()*100/Total_Students_g
Overall_Passing_g=(Passing_Math_g+Passing_Reading_g)/2

summary2=pd.DataFrame({"Total Schools":School_type,
                      "Total Students":Total_Students_g,
                      "Total School Budget":Total_Budget_g.map("\${:,.2f}".format),
                      "Per Student Budget":Student_Budget_g.map("\${:,.2f}".format),
                      "Average Math Score":Average_Math_Score_g,
                      "Average Reading Score":Average_Reading_Score_g,
                      "% Passing Math":Passing_Math_g,
                      "% Passing Reading":Passing_Reading_g,
                      "% Overall Passing Rate":Overall_Passing_g})

summary2.sort_values(by=["% Overall Passing Rate"], ascending=False).head(5)
```
